```python
import dbnomics
import pandas as pd

# Fetch series with a filter for Chinese data
df = dbnomics.fetch_series(provider_code='Eurostat', dataset_code='migr_imm1ctz', dimensions={'citizen': ['CN']}, max_nb_series=2_000_000)

# Convert the data to a pandas DataFrame for easier manipulation
df = pd.DataFrame(df)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
df
```

```python
total = df[df["sex"] == "T"]
total
```

```python
# all data is reported in annual frequency
print(total["@frequency"].value_counts())

# origianl time
print(total["original_period"].value_counts())
```

```python
# check why 2018 has some many data points

# get a summary which countries are included in the data

total[total["original_period"] == "2018"]["geo"]
```

```python
total[total["original_period"] == "2018"]["geo"].value_counts()
```

```python
total[total["original_period"] == "2018" & total["geo"] == "BE"]
```

```python
total[(total["original_period"] == "2018") & (total["geo"] == "BE") & (total["age"] == "TOTAL") & (total["series_name"].str.contains("completed"))]
```

```python
total_clean = total[(total["age"] == "TOTAL") & (total["series_name"].str.contains("completed"))]
```

```python
# provide a line plot for the data
import matplotlib.pyplot as plt

total_clean.plot(x='period', y='value', kind='line')
plt.xlabel('Period')
plt.ylabel('Value')
plt.title('Line Plot of Total Clean')
plt.show()
```
